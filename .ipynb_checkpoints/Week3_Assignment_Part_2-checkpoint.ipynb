Importing libraries and get data from csv (imported as csv in part1)

```python
import pandas as pd
import numpy as np
import geocoder


toronto = pd.read_csv('toronto.csv')
toronto.head()
```

Function for getting lat and lon values

```python
def get_lat_lon(postal_code):
    lat_lon_coords = None
    while(lat_lon_coords is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        lat_lon_coords = g.latlng
    return lat_lon_coords
```

Retrieving postal codes and adding to dataframe

```python
postal_codes = toronto['PostalCode']    
coords = [get_lat_lon(postal_code) for postal_code in postal_codes.tolist()]

df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])

toronto['Latitude'] = df_coords['Latitude']
toronto['Longitude'] = df_coords['Longitude']

toronto.head(20)
```

```python
# import as csv
toronto.to_csv('toronto_w_lat_lon.csv', index=False)
```
